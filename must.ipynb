```python
def process_dataframe_old(df):
    for i in range(0, len(df), 2):
        if i+1 < len(df):
            ls = df['text'].iloc[i].split(" </s> ")
            df['text'].iloc[i+1] = ls[-2].strip() + " </s></s> " + df['text'].iloc[i+1]
            df['text'].iloc[i] = ' </s>'.join(ls[:-2])
            
    return df

def process_dataframe(df):
    for i in range(0, len(df), 2):
        if i+1 < len(df):
            ls = df['text'].iloc[i].split(" </s> ")
            utterance = df['text'].iloc[i+1]
            # Count the number of words in the utterance
            utterance_word_count = len(utterance.split())
            # Calculate the maximum number of words we can take from the context
            max_context_words = 500 - utterance_word_count
            # Join the context utterances into a single string, starting from the end
            context = ''
            for context_utterance in reversed(ls):
                context_utterance_words = context_utterance.split()
                if len(context_utterance_words) <= max_context_words:
                    context = context_utterance + " </s> " + context
                    max_context_words -= len(context_utterance_words)
                else:
                    break
            df['text'].iloc[i+1] = context.strip() + " </s></s> " + utterance
    return df
df = process_dataframe(df)
df['text']
```
