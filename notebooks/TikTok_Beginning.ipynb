```python
root_dirs = ['../data/tiktok_videos/test' ,  '../data/tiktok_videos/val' , '../data/tiktok_videos/train']
subfolders = ['educative', 'others', 'suggestive']
data = []
dc = {'others' : 0 , 'suggestive' : 1 , 'educative' : 2}
for root_dir in root_dirs:
    for subfolder in subfolders:
        path = os.path.join(root_dir, subfolder)
        for root, dirs, files in os.walk(path):
            for file in files:
                if file.endswith('.mp4'):
                    full_path = os.path.join(root, file)
                    text_path = "../data/transcripts/" + full_path.split('../data/tiktok_videos/')[1][:-4] + ".txt"
                    with open(text_path, 'r') as f:
                        text = f.read()
                        data.append([ re.sub(r"_\d+", "", full_path.split('/')[-1][:-4]), subfolder , dc[subfolder] , text , full_path , 
                                     full_path.replace("tiktok_videos" , "tiktok_wav")+".wav" , 
                                     root_dir.split('/')[-1] , len(text)])
                        
# text , audio_path , video_path , label , label_name , split , dialog , timings

df = pd.DataFrame(data, columns=['name' , 'content_label', 'content'  , 'text' , 'video_path' , 'audio_path', 'split' , 'text_len'])
df.loc[df['text_len'] == 0, "text"] = "[N/A]"

df_train = df[df['split'] == 'train']
df_test = df[df['split'] == 'test']
df_val = df[df['split'] == 'val']
df_train['dialog'] = df_train.groupby('name').ngroup()
df_test['dialog'] = df_test.groupby('name').ngroup()
df_val['dialog'] = df_val.groupby('name').ngroup()

df = pd.concat([df_test , df_val , df_train])

df
```

```python
def fn(text):
    try:
        return detect(text)
    except:
        return None


df['language'] = df['text'].apply(fn)
```
