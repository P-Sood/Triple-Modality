```python
import pandas as pd
```

```python
old_df =  pd.read_pickle("data/iemo.pkl")
test_df =  pd.read_pickle("data/IEMO_test_new.pkl")
train_df =  pd.read_pickle("data/IEMO_train_new.pkl")
val_df =  pd.read_pickle("data/IEMO_val_new.pkl")
```

```python
test_df
```

```python
test_df['utterance_order'] = test_df.groupby('index').cumcount()
train_df['utterance_order'] = train_df.groupby('index').cumcount()
val_df['utterance_order'] = val_df.groupby('index').cumcount()
old_df['utterance_order'] = old_df.groupby('index').cumcount()
```

```python
old_df
```

```python
train_df
```

```python
df = pd.concat([train_df , val_df , test_df])
df = df.rename(columns = {"label":"emotion"})
df
```

```python
new_df = old_df.drop(["text" , "concatenated" , "emotion" , "text_w_one_prev" , "og_text", "og_text_with_auth"], axis=1)
new_df
```

```python
merged = pd.merge(df , new_df, how = "inner" , on = ['split' ,'index' , 'utterance_order'])
merged
```

```python
merged.to_pickle("data/more_text_IEMO.pkl")
```
